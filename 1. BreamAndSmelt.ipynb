_**Classification of Bream and Smelt by using k-Nearest Neighbors Algorithm**_



## 1. **Preparing Data of Breams**


```python
# Features of Bream : Length, Weight
# Data set of 35 Breams

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel(bream_length)    # x-axis : length
plt.ylabel(bream_weight)    # y-axis : weight
plt.show()
```

## 2. **Preparing Data of Smelts**

```python
# Features of Bream : Length, Weight
# Data set of 14 Smelts

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel(smelt_length)
plt.ylabel(smelt_weight)
plt.show()
```

#3. **Data Processing for scikit-learn**

```python
# 길이와 무게를 짝지어 2D array 생성 -> 머신러닝 패키지 scikit-learn 사용을 위해
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

fish_data = [[l, w] for l, w in zip(length, weight)]  # zip()이 각 배열에서 원소 하나씩 반환

print(fish_data)
```

#4. **Preparing Label for Data Set**

```python
# Binary Classification에서는 찾고자 하는 대상을 1, 나머지 대상을 0으로 labelling
fish_target = [1] * 35 + [0] * 14     # fish_data의 앞부분 데이터인 도미 35마리를 1, 이어서 오는 빙어 14마리를 0으로 labelling
print(fish_target)
```

#5. **Training and Evaluation using scikit-learn**

```python
from sklearn.neighbors import KNeighborsClassifier   # scikit-learn 패키지의 k-Nearest Neighbor Algorithm을 구현한 클래스

kn = KNeighborsClassifier()

kn.fit(fish_data, fish_target)  # fit() method를 통한 training 실행

kn.score(fish_data, fish_target)  # score() method를 통한 model 평가 -> 0과 1사이의 값 반환

# 1 : 모든 데이터셋의 정답을 맞춤
# 0 : 데이터셋의 정답을 하나도 맞추지 못함
```

# ✅ **k-Nearest Neighbor Algorithm**
> * 어떤 데이터가 주어졌을 때 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 한다
  
>* 데이터가 많은 경우 메모리 소모가 크고 주변 데이터와의 거리계산에 많은 시간 소요
  
>* KNeighborsClassifier() 클래스의 최소거리 주변 데이터 참고의 기본값은 5 이다



```python
kn.predict([[30, 600]])   # 새로운 데이터를 넣어주고 predict() method를 통해 정답 예측
```

```python
print(kn._fit_X)   # KNeighborsCllasifier() 클래스의 _fit_X 속성으로는 fish_data를 가지고 있다
```

```python
print(kn._y)  # _y 속성으로는 label정보인 fish_target을 가지고 있다
```

#❗**가까운 데이터의 참고 기본값 수정가능**

```python
kn49 = KNeighborsClassifier(n_neighbors = 49)  # 주변의 49개의 가까운 값 참고

kn49.fit(fish_data, fish_target)  # 도미와 빙어가 총 49마리이면서 도미가 다수를 차지하기에 어떤 데이터를 넣어도 모두 도미로 판단할 것이다

kn49.score(fish_data, fish_target) # 식별값을 49로 하여 정확도를 계산하면 35/49 가 나온다
```

# ➕ **정확도가 1보다 낮아지는 지점 찾기**

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for i in range(5, 50) :     # 식별값을 늘려가기
  kn.n_neighbors = i
  score = kn.score(fish_data, fish_target)
  if score < 1 :
    print(i, score)
    break
```
